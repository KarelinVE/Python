```python
import json
purchases = {}
        
with open('purchase_log.txt', 'r', encoding = 'utf-8') as file:
    for i, line in enumerate(file):
        line = line.strip()
        dicts = json.loads(line)
        purchases[dicts['user_id']] = dicts['category']
        if i > 25:
            break
print (purchases, '\n \n \n')

f = str()
line_user_id = str()

with open('visit_log.csv', 'r') as visits:
    categories = open('funnel.csv', 'w')
    for i, line in enumerate(visits):
        line = line.strip ('\n')
        line_user = line.split(',')[0]
        line_category = purchases.get(line_user)
        if line_category:
            f = line + ',' + line_category + '\n'
            categories.write(f)
            print(f)
        i += 1
        if i > 1000:
            break
```
