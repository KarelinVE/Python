```python
import pandas as pd

ratings = pd.read_csv('ml-latest-small\\ratings.csv')
power = pd.read_csv('ml-latest-small\\power.csv')
```

```python
#Фильм снаибольшим количеством оценок 5.0

best_films = (ratings.query('rating == 5')
.groupby('movieId').count()
.reset_index()[['movieId', 'rating']]
.sort_values('rating', ascending = False))
best_films.head(1)
```

```python
#Суммарное потребление стран Прибалтики

f = power [((power['country'] == 'Lithuania') | (power['country'] == 'Estonia') | (power['country'] == 'Latvia'))
           & ((power['category'] == 4) | (power['category'] == 12) | (power['category'] == 21))
           & ((power['year'] > 2004) & (power['year'] < 2011))
           & (power['quantity'] > 0)
]

f['quantity'].count()
```
